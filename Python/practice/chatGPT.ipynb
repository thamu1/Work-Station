```python
import openai
```

```python
openai.api_key = '*******************'
```

```python
model_engine = 'text-davinci-003'
```

```python
while(True):
    prompt = input("enter your Qn :")
    if(prompt == "exit" or prompt == "leave"):
        break
    else:
        completion = openai.Completion.create(
            engine = model_engine,
            prompt = prompt,
            max_tokens = 1024,
            n = 1,
            stop = None,
            temperature = 0.5  
            )
        response = completion.choices[0].text
        print(response)
    
```
